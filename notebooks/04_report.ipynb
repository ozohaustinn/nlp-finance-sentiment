# 04 — Comparative Sentiment Report
**Goal:** Visualise and compare FinBERT sentiment scores across JPMorgan Chase and Citigroup Q2 2025 10-Q filings.

**Inputs:**
- `outputs/JPM_sentiment_latest.json`
- `outputs/C_sentiment_latest.json`

**What this report shows:**
1. Overall sentiment score comparison
2. Sentiment distribution (positive / negative / neutral)
3. Category-level scores across both banks
4. Top risk phrases flagged per institution
5. Most negative and most positive sentences per bank

```python
import json
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Plot styling
plt.rcParams.update({
    "figure.facecolor": "white",
    "axes.facecolor": "white",
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "grid.alpha": 0.3,
    "font.family": "sans-serif",
    "font.size": 11,
})

COLORS = {
    "JPM": "#1a1a2e",
    "C":   "#c0392b",
    "positive": "#27ae60",
    "negative": "#e74c3c",
    "neutral":  "#95a5a6",
}

def load_sentiment(ticker):
    path = pathlib.Path(f"../outputs/{ticker}_sentiment_latest.json")
    with open(path) as f:
        return json.load(f)

jpm = load_sentiment("JPM")
cit = load_sentiment("C")

print(f"JPM scored: {jpm['summary']['total_scored']} sentences")
print(f"C   scored: {cit['summary']['total_scored']} sentences")
print("\nLoaded successfully.")
```
